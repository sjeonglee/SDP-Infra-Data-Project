```
# cleaning the text
def clean_text(text):
  tokens = word_tokenize(text)                     # tokenize the words
  tokens = [w.lower() for w in tokens]             # lower the case
  tokens = [w for w in tokens if w.isalpha()]      # remove the punctuations
  tokens = [w for w in tokens if not w in stops]   # remove stopwords
  return tokens

# stemming the tokens if needed
def stemming_the_tokens(tokens):
  snowball = SnowballStemmer("english")
  return [snowball.stem(w) for w in tokens]
```

```
# count the frequency and find duplicates
def word_list_to_frequency(word_list):
    word_freq = [word_list.count(w) for w in word_list]
    return dict(list(zip(word_list, word_freq)))

# remove duplicates of the words in the word_list
def word_list_remove_duplicates(word_list):
  word_set = set(word_list)
  word_list = list(word_set)
  return word_list
```

```
# the dictionary to be a dataframe
tobe_df = {}
# a list of columns of the dataframe
tobe_column = []

for i in range(len(df_example)):
  try:
    current_row = df_example.iloc[i]
    status = current_row['status']
    body = current_row['body']
    tokens = clean_text(body)
    word_freq = word_list_to_frequency(tokens)
    word_list = word_list_remove_duplicates(tokens)
    
    # write the result to a dictionary
    tobe_df[i] = [status, body, word_list, word_freq]

    # add the columns
    tobe_column.extend(word_list)

    # column_list = word_freq.keys()
    # df_example_columns = df_example.columns.tolist()
    # create_on = [x for x in column_list if x in df_example_columns]
  except:
    continue

print(tobe_df)
tobe_column = list(set(tobe_column))
print(tobe_column)
```

```
# compose the final dataframe (columns)
df_recompose = {'status': [], 'body': [], 'word_list': []}
for j in tobe_column:
  df_recompose[j] = []

for i in tobe_df:
  current_pack = tobe_df[i]
  # every elements in the word_freq
  df_recompose['status'].append(current_pack[0])
  df_recompose['body'].append(current_pack[1])
  df_recompose['word_list'].append(current_pack[2])
  print(current_pack[3])
  for j in tobe_column:
    if j in current_pack[3]:
      df_recompose[j].append(current_pack[3][j])
    else:
      df_recompose[j].append(0)
```
